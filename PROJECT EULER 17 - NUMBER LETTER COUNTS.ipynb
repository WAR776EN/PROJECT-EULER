```python
satuan = ['zero','one','two','three','four','five','six','seven','eight','nine']
belasan =['ten','eleven','twelve','thirteen','fourteen','fifteen','sixteen','seventeen','eighteen','nineteen']
puluhan =['twenty','thirty','forty','fifty','sixty','seventy','eighty','ninety']
add = ['thousand','hundred','and']
    
def satu (x):
    global satuan
    global belasan
    global puluhan
    global add
    
    letter = satuan[x]
    result = len(letter)
    return letter

def dua (x):
    global satuan
    global belasan
    global puluhan
    global add
  
    numstr = str(x)
    if numstr[0]=='1':
        letter = belasan[int(numstr[1])]
        result = len(letter)
        return letter
    elif (numstr[0] != '1') and (numstr [1]=='0'):
        letter = puluhan[int(numstr[0])-2]
        result = len(letter)
        return letter
    else:
        letter = puluhan[int(numstr[0])-2] + satu(int(numstr[1]))
        result = len(letter)
        return letter
```

```python
def tiga (x):
    global satuan
    global belasan
    global puluhan
    global add

    numstr = str(x)
    if numstr[1:] == '00':
        letter = satuan[int(numstr[0])] + add[1]
        result = len(letter)
        return letter
    elif numstr[1]=='0' and numstr[2] != '0':
        letter = satuan[int(numstr[0])] + add[1] +add[2]+ satu(int(numstr[2]))
        result = len(letter)
        return letter
    else:
        letter = satuan[int(numstr[0])] + add[1] + add[2] + dua(int(numstr[1:]))
        result = len(letter)
        return letter
```

```python
def empat (x):
    global satuan
    global belasan
    global puluhan
    global add
    
    numstr = str(x)
    if numstr[1:] == '000':
        letter = satuan[int(numstr[0])] + add[0]
        result = len(letter)  
        return letter
    elif numstr[1:3] =='00':
        letter = satuan[int(numstr[0])] + add[0] + add[2] + satu(int(numstr[3]))
        result = len(letter)
        return letter
    elif numstr[1] == '0' and numstr[2:]!='0':
        letter = satuan[int(numstr[0])] + add[0] + add[2] + dua(int(numstr[2:]))
        result = len(letter)
        return letter
    else:
        letter = satuan[int(numstr[0])] + add[0] + tiga(int(numstr[1:]))
        result = len(letter)
        return letter
        
```

```python
def lima(x):
    global satuan
    global belasan
    global puluhan
    global add
    
    numstr = str(x)

    if numstr[2:] == '000':
        letter = dua(int(numstr[0:2])) + add[0]
        result = len(letter)  
        return letter
    elif numstr[2:4] =='00':
        letter = dua(int(numstr[0:2])) + add[0] + add[2] + satu(int(numstr[-1]))
        result = len(letter)
        return letter
    elif numstr[2] == '0' and numstr[3:]!='0':
        letter = dua(int(numstr[0:2])) + add[0] + add[2] + dua(int(numstr[-2:]))
        result = len(letter)
        return letter
    else:
        letter = dua(int(numstr[0:2])) + add[0] + tiga(int(numstr[2:]))
        result = len(letter)
        return letter
```

```python
def nlc(bottom,top, y=False):
    
    count = 0
    bottomPrint = bottom
    while bottom <= top:
        if bottom < 10:
            count += len(satu(bottom))
            if y == True:
                print(satu(bottom))
        if bottom >= 10 and bottom <100:
            count += len(dua(bottom))
            if y == True:
                print(dua(bottom))
        if bottom >=100 and bottom <1000:
            count += len(tiga(bottom))
            if y == True:
                print(tiga(bottom))
        if bottom >=1000 and bottom <10000:
            count += len(empat(bottom))
            if y == True:
                print(empat(bottom))
        if bottom >= 10000 and bottom <100000:
            count += len(lima(bottom))
            if y == True:
                print(lima(bottom))
        bottom += 1
    
    if y==True:
        print(str(bottomPrint)+' to '+str(top)+' used '+ str(count)+' letters in total')
    return count
```
